```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()
```

```python
E_commerce_data = pd.read_csv("C:\\Users\\User\\Documents\\PDS_PROJ_PIC\\ECOMMERCE.csv")
```

```python
E_commerce_data
```

```python
E_commerce_data = E_commerce_data.rename({"Highest Educational Qualification":"Qualification",
                                         "Do you generally find various alternatives for the same product in your affordable range?":"People_views_on_alternative_products_on_their_affordable_buying_range"}, axis = 1)
```

```python
E_commerce_data
```

```python
data = pd.DataFrame(columns=["People_views_on_alternative_products_on_their_affordable_buying_range","subscription","Gender"])
```

```python
data['Gender']=E_commerce_data['Gender']
data['People_views_on_alternative_products_on_their_affordable_buying_range']=E_commerce_data.People_views_on_alternative_products_on_their_affordable_buying_range
data['subscription']=E_commerce_data['Age']
```

```python
data
```

```python
data1=data.groupby(['People_views_on_alternative_products_on_their_affordable_buying_range','Gender']).count()
```

```python
data1
```

```python
data1
```

```python
data2 = pd.DataFrame(columns=["People_views_on_alternative_products_on_their_affordable_buying_range","gender"])
row_list = []
for i,j in data1.index:
    dict1 = {}
    dict1.update({"People_views_on_alternative_products_on_their_affordable_buying_range": i, "gender":j})
    row_list.append(dict1)
df1 = pd.DataFrame(row_list); df1    
```

```python
df1['Affordability']=list(data1.subscription)
```

```python
df1
```

```python
df1['Percentage']=df1.groupby("People_views_on_alternative_products_on_their_affordable_buying_range").transform(lambda x: (x*100)/x.sum())
```

```python
df1
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18.5,10.5, forward = True)
sns.barplot(x = 'People_views_on_alternative_products_on_their_affordable_buying_range', y = 'Percentage',
           data = df1,
           hue = 'gender')
fig.savefig('10.png',facecolor='lightgrey',transparent=True)
```
