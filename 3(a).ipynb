```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()
```

```python
E_commerce_data=pd.read_csv("C:\\Users\\User\\Documents\\PDS_PROJ_PIC\\ECOMMERCE.csv")
```

```python
E_commerce_data
```

```python
E_commerce_data = E_commerce_data.rename({"Highest Educational Qualification":"Qualification",
                                         "Which e-commerce sites you prefer for buying grocery items?":"Grocery_sites"}, axis = 1)
```

```python
E_commerce_data
```

```python
E_commerce_data=E_commerce_data.replace('Jio mart',np.nan)
E_commerce_data=E_commerce_data.replace('Future Group',np.nan)
E_commerce_data=E_commerce_data.replace('Future group',np.nan)
E_commerce_data=E_commerce_data.replace('None',np.nan)
E_commerce_data=E_commerce_data.replace('None,',np.nan)
E_commerce_data=E_commerce_data.replace('More',np.nan)
E_commerce_data=E_commerce_data.replace('No one',np.nan)
E_commerce_data=E_commerce_data.replace("I don't buy groceries online","Prefer not to buy")



```

```python
a = E_commerce_data[E_commerce_data['Age']<=19]
a = pd.concat([a]*round(100/len(a)))
b = E_commerce_data[E_commerce_data['Age']<=22][E_commerce_data['Age']>19]
b = pd.concat([b]*round(100/len(b)))
c = E_commerce_data[E_commerce_data['Age']<25][E_commerce_data['Age']>=22]
c = pd.concat([c]*round(100/len(c)))
d = E_commerce_data[E_commerce_data['Age']>=25]
d = pd.concat([d]*round(100/len(d)))
              
```

```python
f = plt.figure()
plt.subplot(321)
a.groupby(['Grocery_sites',])['Grocery_sites'].count().plot.bar(title="Site Preference For Purchasing Grocery Products w.r.t Age <=19",
                                                                                 color = ("red", "yellow", "green","brown", "orange"),
                                                                            figsize = (16,10))
plt.ylabel("Percentage of Grocery Product Purchased")
plt.subplot(322)
b.groupby(['Grocery_sites',])['Grocery_sites'].count().plot.bar(title="Site Preference For Purchasing Grocery Products w.r.t  19<Age<=22",
                                                                                 color = ("red", "yellow", "green","brown", "orange"),
                                                                            figsize = (16,10))
plt.ylabel("Percentage of Grocery Product Purchased")


plt.subplot(325)
c.groupby(['Grocery_sites'])['Grocery_sites'].count().plot.bar(title=" Site Preference For Purchasing Grocery Products w.r.t  22<Age<=25",
                                                                                 color = ("red", "yellow", "green","brown", "orange"),
                                                                            figsize = (16,10))
plt.ylabel("Percentage of Grocery Product Purchased")


plt.subplot(326)
d.groupby(['Grocery_sites',])['Grocery_sites'].count().plot.bar(title="Site Preference For Purchasing Grocery Products w.r.t Age>25",
                                                                                 color = ("red", "yellow", "green","brown", "orange"),
                                                                            figsize = (16,10))
plt.ylabel("Percentage of Grocery Product Purchased")


f.savefig('3(a).png',facecolor='lightgrey',transparent=True)
```
