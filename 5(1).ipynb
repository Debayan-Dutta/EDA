```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
sns.set_theme()
```

```python
E_commerce_data = pd.read_csv("F:\\@BDA\\PDS\\PROJECT\\SURVEY_DATA_ON_ECOMMERCE_SITES.csv")
```

```python
E_commerce_data
```

```python
E_commerce_data = E_commerce_data.rename({"On an average, yearly (approximately), how many products you bought from E-commerce sites on pre-Covid situation?":"Pre",
                                         "On an average, yearly (approximately), how many products you buy from E-commerce sites on post-Covid situation?":"Post",
                                         }, axis = 1)
```

```python
fig = plt.figure()
fig.set_size_inches(18.5,10.5, forward = True)
ax1 = fig.add_subplot(121)
plt.boxplot(E_commerce_data.Pre)
ax1.set_ylim(0,20)
ax2 = fig.add_subplot(122)
plt.boxplot(E_commerce_data.Post)
ax2.set_ylim(0,20)
fig.savefig('5(1).png',facecolor='lightgrey',transparent=True)
```
